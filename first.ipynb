Pulling in data from a public API below

```python
import pandas as pd 
import requests
url = 'https://api.covid19api.com/summary'
r = requests.get(url)
```

The data request worked. Next view object using keys method.

```python
json = r.json()
json.keys()
```

Creating the dataframe

```python
data = r.json()['Countries']
df = pd.DataFrame(data=data)
df.head()
```

Finding results for only Angola

```python
df[df['Country'] == 'Angola']
```

Next focusing on Total Confirmed. Getting the values and then the total sum

```python
df['TotalConfirmed'].value_counts
```

```python
df['TotalConfirmed'].sum()
```

Next using .query() to show all of the countries with more deaths than 60000

```python
df.query('TotalDeaths > 60000')
```

```python
df[df.columns[1:3]]
```
